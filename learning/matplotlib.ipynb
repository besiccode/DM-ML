```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
a = np.linspace(0,10,30)
b = np.sin(a)
c = np.cos(a)

fig = plt.figure()

plt.title("First picture")
plt.xlabel('sin(x)')
plt.ylabel("cos(x)")
plt.plot(a,b,'-h',color='orange',label='sin(x)',markersize=10,linewidth=5,markeredgecolor='red')
plt.plot(a,c,'--',color='black',label='cos(x)')
plt.legend(loc=1) #z


fig.savefig("/Users/mac/Desktop/first.png")
```

```python
plt.plot(a,b,'-d',color='orange',label='sin(x)')
plt.plot(a,c,'--',color='black',label='cos(x)')
plt.xlim(2,6)
plt.ylim(0.25,1.00)
```
